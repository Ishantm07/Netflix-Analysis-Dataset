## Q2. In which year highest number of the TV Shows & Movies were released? Show with bar graph

```python
df.dtypes
```

```python
df['Date_N'] = pd.to_datetime(df.Release_Date)
```

```python
df.head(5)
```

```python
df['Date_N'].dt.year.value_counts()
```

```python
plt.figure(figsize=(5, 3)) 
ax = df['Date_N'].dt.year.value_counts().plot(kind= 'bar')
for bar in ax.containers:
    ax.bar_label(bar)
plt.show()
```
